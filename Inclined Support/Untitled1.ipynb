# 3D truss with inclined support
We are about to solve the following problem assuming that the support is inclined at node 5 with a 45 degree angle with each axises of x , y and z:

![](Inclined.png)

```python
#importing necesseties
import numpy as np
import sympy as sp
from sympy import Matrix , symbols, Symbol , zeros as spzeros , N
```

```python
cx , cy ,cz = symbols("Cx Cy Cz")
位 = np.array([[cx**2 , cx*cy , cx*cz],
                [cy*cx , cy**2 , cy*cz],
                [cz*cx , cz*cy , cz*cz]])
位 = Matrix(位)
位1, 位2, 位3 , 位4 = symbols("位1 位2 位3 位4")
```

### Element 1-2

```python
subs = {cx : -0.8, cy : 0 , cz :-0.6}
位1 = np.array([[cx**2 , cx*cy , cx*cz],
                [cy*cx , cy**2 , cy*cz],
                [cz*cx , cz*cy , cz*cz]])
位1 = Matrix(位1)
位1 = N(位1.subs(subs))
k1 = Matrix(15,15,lambda i,j:0)
k1[0:3,0:3] = 位1
k1[0:3,3:6] = -位1
k1[3:6,0:3] = -位1
k1[3:6,3:6] = 位1
K1 = 42000 * k1
```

### Element 1-3

```python
subs = {cx : -0.8, cy : 0 , cz :0.6}
位2 =np.array([[cx**2 , cx*cy , cx*cz],
                [cy*cx , cy**2 , cy*cz],
                [cz*cx , cz*cy , cz*cz]])
位2 = Matrix(位2)
位2 = N(位2.subs(subs))
k2 = Matrix(15,15,lambda i,j:0)
k2[0:3,0:3] = 位2
k2[0:3,6:9] = - 位2
k2[6:9,0:3] = -位2
k2[6:9,6:9] = 位2
K2 = 42000 * k2
```

### Element 1-4

```python
subs = {cx : 0, cy : -1 , cz : 0}
位3 =np.array([[cx**2 , cx*cy , cx*cz],
                [cy*cx , cy**2 , cy*cz],
                [cz*cx , cz*cy , cz*cz]])
位3 = Matrix(位3)
位3 = N(位3.subs(subs))
k3 = Matrix(15,15,lambda i,j:0)
k3[0:3,0:3] = 位3
k3[0:3,9:12] = - 位3
k3[9:12,0:3] = -位3
k3[9:12,9:12] = 位3
K3 = 52500 * k3
```

### Element 1-5

```python
subs = {cx : 0.596, cy : -0.745 , cz : -0.298}
位4 =np.array([[cx**2 , cx*cy , cx*cz],
                [cy*cx , cy**2 , cy*cz],
                [cz*cx , cz*cy , cz*cz]])
位4 = Matrix(位3)
位4 = N(位4.subs(subs))
k4 = Matrix(15,15,lambda i,j:0)
k4[0:3,0:3] = 位4
k4[0:3,12:15] = - 位4
k4[12:15,0:3] = -位4
k4[12:15,12:15] = 位4
K4 = 31305 * k4
```

### Transformation Matrix
Now we need a 15x15 transformation matix for element 4 to compencate the inclined support at node 5

```python
T1 = Matrix(15,15,lambda i,j:0)
from math import sqrt
t4 = Matrix(3,3,[sqrt(2)/2,sqrt(2)/2,sqrt(2)/2,sqrt(2)/2,sqrt(2)/2,sqrt(2)/2,sqrt(2)/2,sqrt(2)/2,sqrt(2)/2])
T1[0:12,0:12] = Matrix(12,12,lambda i,j:1 if i==j else 0)
T1[12:15,12:15] = t4
#The transformation matrix is deriven (T1)
#Now we need to derive the transposition of the T1 ([]) to obtain [T][Kg][]

```

```python
T2 = np.transpose(T1)
T2
```

```python
Kg = K1 + K2 + K3 + K4 
```

```python
K = T1 * Kg * T2
K
```

Applying the boundary conditions where all deflections at node 2, 3, 4 and 5 are zero.
